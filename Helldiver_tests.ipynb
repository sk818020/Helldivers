```python
import json
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

res = requests.get("https://helldiverstrainingmanual.com/api/v1/war/campaign")
response = json.loads(res.text)
df = pd.DataFrame(response)
#df.sort_values(by=['players'], ascending = False).head(20)
df.groupby(['name']).agg({'players':'sum'}).sort_values(by='players', ascending = False).reset_index().plot(kind = 'bar', x = 'name', y = 'players');
```
